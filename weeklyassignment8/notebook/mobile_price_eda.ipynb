Mobile range

```python
sns.set()
price_plot=mobile_data['price_range'].value_counts().plot(kind='bar')
plt.xlabel('price_range')
plt.ylabel('Count')
plt.show()
```

Battery power

```python

sns.set(rc={'figure.figsize':(5,5)})
ax=sns.displot(mobile_data["battery_power"])
plt.show()
```

Bluetooth

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.barplot(data=mobile_data,x='blue',y='price_range',ax=ax)
```

Ram vs price here we can see as price increase with ram

```python
mobile_data.plot(x='price_range',y='ram',kind='scatter')
plt.show()
```

price vs front camera so  according to graph fc is not most affective factor for prediction

```python

mobile_data.plot(x='price_range',y='fc',kind='scatter')
plt.show()
     
```

Mobile weight is affective here as price increase weight is lighter

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
sns.kdeplot(data=mobile_data, x='mobile_wt', hue='price_range', ax=axs[0])
sns.boxplot(data=mobile_data, x='price_range', y='mobile_wt', ax=axs[1])
plt.show()
```

```python
corr = mobile_data.corr()
corr = corr.price_range
cr = corr.sort_values(ascending = False)[1:]
sns.barplot(x=cr, y=cr.index,palette = "bright")
plt.title("Correlation between Attributes and Price Range")
```
